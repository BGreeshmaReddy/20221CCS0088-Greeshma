```python
import folium
mymap=folium.Map(location=[12.978889, 77.591667],
                 width=950,
                 height=550,
                 zoom_start=12,
                 tiles='openstreetmap')
folium.Marker(location=[12.978889, 77.591667],popup='Default popup Marker1',tooltip='Click here to see Popup').add_to(mymap)
folium.Marker(location=[12.9259, 77.6229],popup='<strong>Marker3</strong>',tooltip='<strong>Click here to see Popup</strong>').add_to(mymap)
folium.Marker(location=[12.92, 77.61],popup='<h3 style="color:green;">Marker2</h3>',tooltip='<strong>Click here to see Popup</strong>').add_to(mymap)



mymap
```

```python
import folium


m = folium.Map(location=[12.978889, 77.591667], zoom_start=13)


locations = [[12.978889, 77.591667], [12.9259, 77.6229], [13.23, 77.7],[13.113889, 77.598333],[13.43, 77.72]]
image_paths = [
    "https://www.godrej-ananda.net.in/images/other/how-bangalore-got-its-name.webp",
    "https://imgmedia.lbb.in/media/2022/01/61efa39514a08c2400a7ce56_1643094933277.png",
    "https://assets-news.housing.com/news/wp-content/uploads/2016/11/23154121/Devanahalli.jpg",
    "https://leopackersandmovers.com/intercity/extra-images/leo-packers-movers-yelahanka-location-image.png",
    "https://media-cdn.tripadvisor.com/media/photo-c/1280x250/0c/e9/ec/53/nandi-hills.jpg"

]


for location, image_path in zip(locations, image_paths):
  html = f'<img src="{image_path}" alt="Picture" width="200">'
  folium.Marker(location, popup=html).add_to(m)

m
```
